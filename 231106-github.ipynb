```python
issues_final = pd.read_pickle('/Users/selincaglar/Documents/IS-MSc/Thesis/github_data/231008-all-files/231008-issues-final-nlp.pkl')
```

```python
issues_final = issues_final.drop(issues_final[issues_final['issue_resolution_time'] < 0].index)
issues_final = issues_final.reset_index(drop=True)
```

```python
i = 0
while i < releases.shape[0]:
    repository = releases.loc[i,'repository_name']
    release_tag = releases.loc[i,'tag_name']
    issues = issues_final[issues_final['repository_name'] == repository]
    issues = issues[issues['issue_release_tag_name'] == release_tag]
    releases.loc[i,'#_of_issues'] = issues.shape[0]
    issues = issues[issues['state'] == 'closed']
    releases.loc[i,'avg_resolution_time'] = issues['issue_resolution_time'].mean()
    i = i + 1
```
